```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load the CSV data into a Pandas DataFrame
data = pd.read_csv('database.csv', sep=';')
```

```python
data
```

```python
# 1. Column chart displaying data grouped by type
type_counts = data['type'].value_counts()
type_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Distribution of Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Chart of types per year
year_type_counts = data.groupby(['year', 'type']).size().unstack('type', fill_value=0)
year_type_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Types per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Type')
plt.show()
```

```python
# 3. Chart with publishers
publisher_counts = data['Publisher'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=publisher_counts.index, y=publisher_counts.values)
plt.title('Publisher Distribution')
plt.xlabel('Publisher')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Extract keywords and create a list of individual keywords
all_keywords = []
for keywords in data['keywords']:
    if pd.notnull(keywords):
        for keyword in keywords.split(';'):
            all_keywords.append(keyword.strip().lower())

# Convert the list of keywords to a Pandas Series
keyword_series = pd.Series(all_keywords)

# Create a DataFrame with keyword counts
keyword_counts = keyword_series.value_counts().reset_index()
keyword_counts.columns = ['Keyword', 'Count']

# Create the keyword histogram
plt.figure(figsize=(10, 6))
sns.barplot(data=keyword_counts, x='Keyword', y='Count')
plt.title('Keyword Histogram')
plt.xlabel('Keyword')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Function to create a word cloud with a minimum occurrence and a maximum number of words
def create_word_cloud(text, title, min_occurrence=2):
    # Convert text to lowercase and split into words
    words = ' '.join(text).lower().split()
    
    # Create a frequency dictionary
    freq_dict = {}
    for word in words:
        freq_dict[word] = freq_dict.get(word, 0) + 1
    
    # Filter words based on minimum occurrence
    filtered_words = [word for word, count in freq_dict.items() if count >= min_occurrence]
    
    # Sort the filtered words by frequency in descending order
    sorted_words = sorted(filtered_words, key=lambda x: freq_dict[x], reverse=True)
    
    # Join the filtered words back into a single string
    filtered_text = ' '.join(sorted_words)
    
    # Create and display the word cloud
    wordcloud = WordCloud(background_color='white', width=800, height=400).generate(filtered_text)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title)
    plt.show()
```

```python
# Word cloud for titles
title_text = data['title'].str.lower()
create_word_cloud(title_text, 'Word Cloud for Titles', min_occurrence=2)
```

```python
# Word cloud for abstracts
abstract_text = data['abstract'].str.lower()
create_word_cloud(abstract_text, 'Word Cloud for Abstracts', min_occurrence=3)
```
